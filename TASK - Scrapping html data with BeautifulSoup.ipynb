```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = input ('ENTER - ')
html = urlopen(url).read()
soup = BeautifulSoup(html, 'html.parser')
```

```python
#showing the different things we can do with/from the tag
tags = soup('span')

for tag in tags:
    print('TAG:', tag)
    print (tag.get('class', None))
    print('Contents:', tag.contents[0])
    print('Attrs:', tag.attrs)
```

```python
chris = []
for tag in tags:
    num = tag.contents[0]
    chris.append(num)
chris
```

```python
summ = []
for i in chris:
    i = int (i)
    summ.append(i)
sum(summ)
```
